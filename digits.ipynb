```python
import numpy as np
from tensorflow.keras.datasets import mnist
```

```python

(X, Y), (x, y) = mnist.load_data()

X = X.reshape(-1, 784) / 255.0
x = x.reshape(-1, 784) / 255.0

Y_test = np.zeros(shape=(Y.shape[0], 10))
for i in enumerate(Y):
    Y_test[i[0]][int(i[1])] = 1
Y = Y_test

y_test = np.zeros(shape=(y.shape[0], 10))
for i in enumerate(y):
    y_test[i[0]][int(i[1])] = 1
y = y_test
```

```python
W = []
b = []
```

```python
W.append(np.random.uniform(low=-(np.sqrt(6/784)), high=(np.sqrt(6/784)), size=(784, 20)))
W.append(np.random.uniform(low=-(np.sqrt(6/20)), high=(np.sqrt(6/20)), size=(20, 10)))

b.append(np.zeros(shape=(1, 20)))
b.append(np.zeros(shape=(1, 10)))
```

```python
def softmax(z):
    exp = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp / np.sum(exp, axis=1, keepdims=True)

def relu(z):
    return np.maximum(0, z)

def relu_d(z):
    return z>0
```

```python
def forward(x):
    return softmax((relu((x @ W[0]) + b[0]) @ W[1]) + b[1])
```

```python
for _ in range(25):
    for i in range(X.shape[0]//200):
        A0 = X[i*200:(i+1)*200]
        Z1 = (A0 @ W[0]) + b[0]
        A1 = relu(Z1)
        Z2 = (A1 @ W[1]) + b[1]
        A2 = softmax(Z2)

        dZ2 = A2 - Y[i*200:(i+1)*200]
        dW2 = (1/200) * (A1.T @ dZ2)
        db2 = (1/200) * np.sum(dZ2, axis=0, keepdims=True)
        dA1 = dZ2 @ W[1].T

        dZ1 = dA1 * relu_d(Z1)
        dW1 = (1/200) * (A0.T @ dZ1)
        db1 = (1/200) * np.sum(dZ1, axis=0, keepdims=True)

        W[0] = W[0] - (0.05 * dW1)
        W[1] = W[1] - (0.05 * dW2)

        b[0] = b[0] - (0.05 * db1)
        b[1] = b[1] - (0.05 * db2)

    print(f"epoch {_}")
        
```

```python
result = np.argmax(forward(x), axis=1)
act = np.argmax(y, axis=1)
```
